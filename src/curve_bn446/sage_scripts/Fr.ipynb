```python
def print_repr(x):
    a1 = x % 2^64
    x = x - a1
    x /= 2^64
    a2 = x % 2^64
    x = x - a2
    x /= 2^64
    a3 = x % 2^64
    x = x - a3
    x /= 2^64
    a4 = x % 2^64
    x = x - a4
    x /= 2^64
    a5 = x % 2^64
    x = x - a5
    x /= 2^64
    a6 = x % 2^64
    x = x - a6
    x /= 2^64
    a7 = x
    print("{}u64,".format(a1))
    print("{}u64,".format(a2))
    print("{}u64,".format(a3))
    print("{}u64,".format(a4))
    print("{}u64,".format(a5))
    print("{}u64,".format(a6))
    print("{}u64,".format(a7))
print(n)
print_repr(n)
```
